```python

for i in range(1,51):
    link = lambda x : f"https://books.toscrape.com/catalogue/page-{x}.html"
    soup = BeautifulSoup(requests.get(link(i)).text)
    link1 = soup.find_all("li" ,class_ = "col-xs-6 col-sm-4 col-md-3 col-lg-3")
    baseurl = "https://books.toscrape.com/catalogue/"
    books = []
    for i in link1:
        A = i.find("a",href = True)["href"]
        books.append(requests.get(f"{baseurl}{A}").text)

    book = {}
    soups = list(map(lambda x : BeautifulSoup(x), books))
    name = lambda x : x.find("h1").text
    price = lambda x:x.find("p").text
    ID = lambda x:x.find_all("td")[0].text
    Reviews = lambda x:x.find_all("td")[-1].text
    Genre = lambda x:x.find_all("a")[-1].text
    Ratings = lambda x:x.select_one('.star-rating').attrs["class"][1]
    

    book["name"] = list(map(name,soups))
    book["price"] = list(map(price,soups))
    book["ID"] = list(map(ID,soups))
    book["Reviews"] = list(map(Reviews,soups))
    book["Genre"] = list(map(Genre,soups))
    book["Ratings"] = list(map(Ratings , soups))
    X = pd.DataFrame(book)
    df = pd.concat([df,X])

```
